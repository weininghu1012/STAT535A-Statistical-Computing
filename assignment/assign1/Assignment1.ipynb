```python
import random
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
'''
Question 3: Write pseudo-code to simulate from a mixture 
distribution with 2 normal components
'''
def generateMixtureOfNormals(N, mu1, sigma1, p1, mu2, sigma2, p2):
    samples = np.zeros((1, N))
    uniform = np.random.uniform(0, 1, N)
    for i in range(N):
        if (uniform[i] < p1):
            samples[0][i] = np.random.normal(mu1, sigma1, 1)
        else:
            samples[0][i] = np.random.normal(mu2, sigma2, 1)
    return np.squeeze(samples)
            
```

```python
testSample1 = generateMixtureOfNormals(100, 1, 1, 0.5, 5, 1, 0.5)
sns.kdeplot(testSample1, bw=0.5)
plt.show()
```

```python
testSample2 = generateMixtureOfNormals(100, 1, 3, 0.3, 100, 10, 0.5)
sns.kdeplot(testSample2, bw=0.5)
plt.show()
```
